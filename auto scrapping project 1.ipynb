```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=headphones"
wanted=['999','2 Pack-Apple Earbuds [Apple MFi Certified] Earphones Wired with Microphone for 3.5mm iPhone Headphones (Built-in Microphone & Volume Control) Compatible with iPhone/iPad/iPod/PC/MP3/4',]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
url1="https://www.amazon.in/s?i=electronics&bbn=1805560031&rh=n%3A976419031%2Cn%3A%21976420031%2Cn%3A1389401031%2Cn%3A1389432031%2Cn%3A1805560031%2Cp_6%3AA14CZOWI0VEHLG%7CA1P3OPO356Q9ZB%7CA2HIN95H5BP4BL%2Cp_89%3AApple&ref=mega_elec_s23_1_2_1_6"
wanted1=["71,999","Apple iPhone 14 (128 GB) - Midnight"]
```

```python
scraper=AutoScraper()
result1=scraper.build(url1,wanted1)
print(result1)
```

```python
scraper.get_result_similar(url1,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_cfai':"Title"})
scraper.keep_rules(['rule_cfai'])
scraper.save('amazon-search')
```

```python
scraper.get_result_similar('https://www.amazon.in/s?k=mi+phones',group_by_alias=True)
```

```python
url="https://www.amazon.in/s?k=iphone"
iphone_list=["60,900","Apple iPhone 13 (128GB) - (Product) RED","15,306"]
```

```python
scraper=AutoScraper()
result=scraper.build(url,iphone_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_t2wp':"Title",'rule_ja0v': 'Price'})
scraper.keep_rules(['rule_t2wp','rule_ja0v'])
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=mi+phone',group_by_alias=True)
```
